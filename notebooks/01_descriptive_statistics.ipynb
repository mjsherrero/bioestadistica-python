```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Sample health dataset
data = {
    "Age": [22, 25, 30, 35, 40, 45, 50, 55],
    "BMI": [21.5, 24.0, 26.1, 27.8, 29.0, 30.2, 31.5, 33.0],
    "Blood_Pressure": [110, 115, 120, 125, 130, 135, 140, 145]
}

df = pd.DataFrame(data)

# Descriptive statistics
print(df.describe())
```

```python
# BMI distribution
plt.hist(df["BMI"])
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.title("BMI Distribution")
plt.show()
```
